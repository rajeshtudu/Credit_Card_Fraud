# Input Split

```python
X = df.drop(columns=['Class'], axis=1)
y = df['Class']
```

# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaler = sc.fit_transform(X)
```

```python
x_scaler[-1]
```

# Model Training

```python
# train test split
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
x_train, x_test, y_train, y_test = train_test_split(x_scaler, y, test_size=0.25, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

# training
model.fit(x_train, y_train)

# testing
y_pred = model.predict(x_test)

# Classification report
print(classification_report(y_test, y_pred))

# f1_score
print('f1_score : ',f1_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()

# training
model.fit(x_train, y_train)

# testing
y_pred = model.predict(x_test)

# Classification report
print(classification_report(y_test, y_pred))

# f1_score
print('f1_score : ',f1_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()

# training
model.fit(x_train, y_train)

# testing
y_pred = model.predict(x_test)

# Classification report
print(classification_report(y_test, y_pred))

# f1_score
print('f1_score : ',f1_score(y_test, y_pred))
```

# Class Imbalancement

```python
sns.countplot(data = X, x = y_train)
plt.show()
```

```python
# hint - use combination of over sampling and under sampling
# balance the class with equal distribution
from imblearn.over_sampling import SMOTE
over_sample = SMOTE()
x_smote, y_smote = over_sample.fit_resample(x_train, y_train)
```

```python
sns.countplot(data = X, x = y_smote)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

# training
model.fit(x_smote, y_smote)

# testing
y_pred = model.predict(x_test)

# Classification report
print(classification_report(y_test, y_pred))

# f1_score
print('f1_score : ',f1_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_jobs = -1)

# training
model.fit(x_smote, y_smote)

# testing
y_pred = model.predict(x_test)

# Classification report
print(classification_report(y_test, y_pred))

# f1_score
print('f1_score : ',f1_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(n_jobs = -1)

# training
model.fit(x_smote, y_smote)

# testing
y_pred = model.predict(x_test)

# Classification report
print(classification_report(y_test, y_pred))

# f1_score
print('f1_score : ',f1_score(y_test, y_pred))
```
